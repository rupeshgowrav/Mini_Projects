# File Handling in python :

### File Handling:
   - Files are used to store the data permenently.
   - Using file handling , we can read,write and modify files stored on disk/memory.
   - python provides built-in funtions and modules to work with files.

### Opening and closing Files:
   - Python provides the open() function to open files.
   - syntax:
    file = open("filename", "mode")

### Closing a file:
  - Always close a file after performing operations to free system resources.
  - syntax: 
     file.close()

### File modes:

- w --> write
- r --> read
- a --> append
- rb --> read in binary mode 
- wb --> write in binary mode 
- ab --> append in binary mode 

### File properties:

- 1.name : name of opened file
- 2.mode : to find mode of the file
- 3.readable() --> T/F
- 4.writable() --> T/F
- 5.closed --> T/F

```python
f = open('abc.txt','w')
print('File name : ',f.name)
print('File mode : ',f.mode)
print(f.readable())
print(f.writable())
print(f.closed)
f.close()
print(f.closed)
```

### The with Statement:
 - with statement automatically closes the file after use.
  - syntax:

           with open("example.txt","r") as file:
              data = file.read()

### Reading and Writing files:

python provides various methods to read and write files 

### Reading a file :

To read the data from file we have following methods:
- read() - to read the complete data 
- read(n) - to read n no.of characters
- readlines() - to read 1st line of the file
- readlines() - to read all lines into list 

```python
with open('abc.txt','r') as file:
    content= file.read()
print(content)
```

```python
with open('abc.txt','r') as file:
    content= file.read(10)
print(content)
```

```python
with open('abc.txt','r') as file:
    content= file.readline()
print(content)
```

```python
with open('abc.txt','r') as file:
    content= file.readlines()
print(content)
```

#### Writing to a file:
To write data into files we use following methods:
- 1.write(): To write string data 
- 2.writelines() : To write list data

### Note:

This erases exisiting content.

```python
with open("abc.txt","w") as file:
    file.write("Aditya Degree College")
```

```python
l=["Aditya Degree College\n","Kakinada\n","Andhra Pradesh"]
with open("abc.txt","w") as file:
    file.writelines(l)
```

```python
with open("abc.txt","a") as file:
    file.write("533003")
```
