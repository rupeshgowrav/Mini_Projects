# Copies and viewing :


```python
import numpy as np
a = np.array([1,2,3,4]) # one dimension array
```

```python
a = np.arange(10)
b = a[::2]
print(a)
print(b)
```

```python
print(np.shares_memory(a,b))
```

```python
b[0] = 10
print(a)
print(b)
```

```python
a = np.arange(10)
b = a[::2].copy() #force a copy
print(a)
print(b)
```

```python
print(np.shares_memory(a,b))
```

# Random number Generator:

```python
a = np.random.randint(0,20,20)
print(a)
```

```python
mask=(a%2==0)
```

```python
mask_a = a[mask]
print(mask_a)
```

```python
a[mask] = -1
print(a)
```

# Element Wise Operations :

## 1. Basic Operation 

```python
a = np.array([1,2,3,4])
print(a+1) #add + 1
```

```python
print(a**2) #squares
```

## 2. All Arthimetic operators Element Wise :

```python
b = np.ones(4) + 1     #[2 2 2 2]
print("a = ",a)
print("b = ",b)
```

```python
print(a*b)
```

```python
print(a-b)
```

```python
# matrix Multiplication
c = np.diag([1,2,3,4])
print(c)
```

```python
print(c*c)
print("***************")
print(c.dot(c))
```

## 3. Comparision Operators :

```python
a = np.array([1,2,3,4])
b = np.array([5,2,7,4])
print(a==b)
```

```python
print(a>b)
```

```python
print(a<b)
```

```python
# Array wise comparison
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
c = np.array([1,2,3,4])
print(np.array_equal(a,b))
```

```python
print(np.array_equal(a,c))
```

## 4. Logical Operators:

```python
a = np.array([1,0,1,0],dtype = "bool")
b = np.array([1,1,0,0],dtype = "bool")
```

```python
print(np.logical_or(a,b))
```

```python
print(np.logical_and(a,b))
```
